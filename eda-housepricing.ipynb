Category frequency

```python
def make_frequency(index):
    vc = dtrain[col_category[index]].value_counts().reset_index()
    vc.columns = ['category', 'frequency']
    
    return vc
```

```python
fig = plt.figure(figsize=(20, 38))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for i in range(1, len(col_category)+1):
    ax = fig.add_subplot(9, 5, i)
    vc = make_frequency(i-1)
    sns.barplot(x = 'category', y = 'frequency', data = vc)
    plt.xlabel(col_category[i-1])
    
plt.savefig('category_frequency.jpg')
```
